```python
import requests
from bs4 import BeautifulSoup  # 引入BS库
import requests
import pandas as pd
import json
import time
```

```python
def get_cex_list(page):
    url = '''https://dncapi.bqrank.net/api/v2/exchange/web-exchange?token=&page={PAGE}&pagesize=100&sort_type=exrank&asc=1&isinnovation=1&type=all&area=&webp=1'''.format(PAGE=page)
    strhtml = requests.get(url) #Get方式获取网页数据
    content = strhtml.text
    result = json.loads(content)
    data = result['data']
    cex = pd.DataFrame(data)
    return cex

#https://dncapi.bqrank.net/api/v2/exchange/web-exchange?token=&page=1&pagesize=100&sort_type=exrank&asc=1&isinnovation=1&type=all&area=&webp=1
#https://dncapi.bqrank.net/api/v2/exchange/web-exchange?token=&page=2&pagesize=100&sort_type=exrank&asc=1&isinnovation=1&type=all&area=&webp=1
```

```python
cex=get_cex_list(1)
time.sleep(3)
for i in range(6):
    page=i+2
    print(page)
    time.sleep(1)
    cex=cex.append(get_cex_list(page))
    
cex.to_csv('cex.csv') 
```

```python
# 非小号排名	rank
# ID	platform
# 交易所名称	platform_name
# logo	logo
# 成立时间	launchedtime
# 注册地区code	countrycode
# 注册地区	country
# 官网地址	official_url
# 备用地址	alternate_url
# 资产实力	assets
# 币种数量	codenum
# 交易对数量	pairnum
# 是否支持KYC	auth_kyc
# 支持中文	supportchinese
# 手续费用	fee_url
# 风险准备金	riskreserve
# 量化接口	quantization
# 人气指数	hotindex
# 介绍	desc
# 交易支持ID	labels_id
# 交易支持	labels
# 社交账号	
# 	weibo_url
# 	twitter_url
# 	facebook_url
# 	telegramlink
# 	redditlink

#https://dncapi.bqrank.net/api/exchange/web-exchangeinfo?code=okex&webp=1
def get_cex_indo(code):
    url = '''https://dncapi.bqrank.net/api/exchange/web-exchangeinfo?code={CODE}&webp=1'''.format(CODE=code)
    strhtml = requests.get(url) #Get方式获取网页数据
    content = strhtml.text
    result = json.loads(content)
    data = result['data']
    code = result['code']
    return code,data


```

```python
code,info=get_cex_indo('bijieex')
if code == 200:
    print(info)
```

```python
csv_path = "./cex_info.csv"
cex_info=pd.read_csv(csv_path, header=0)

s1 = pd.Series(cex["id"]) 
for i, v in s1.iteritems():  
    code,info=get_cex_indo(v)
    time.sleep(1)
    if code == 200:
        print('【已完成】index: ', i, 'value: ', v)
        a = {"rank":info["rank"],
            "platform":info["platform"],
            "platform_name":info["platform_name"],
            "exchange_coinlogo":info["exchange_coinlogo"],
            "launchedtime":info["launchedtime"],
            "countrycode":info["countrycode"],
            "country":info["country"],
            "official_url":info["official_url"],
            "alternate_url":info["alternate_url"],
            "assets":info["assets"],
            "codenum":info["codenum"],
            "pairnum":info["pairnum"],
            "auth_kyc":info["auth_kyc"],
            "supportchinese":info["supportchinese"],
            "fee_url":info["fee_url"],
            "riskreserve":info["riskreserve"],
            "quantization":info["quantization"],
            "hotindex":info["hotindex"],
            "desc":info["desc"],
            "labels_id":info["labels_id"],
            "labels":info["labels"],
            "weibo_url":info["weibo_url"],
            "twitter_url":info["twitter_url"],
            "facebook_url":info["facebook_url"],
            "telegramlink":info["telegramlink"],
            "redditlink":info["redditlink"]}
        cex_info=cex_info.append(a,ignore_index=True)
    else:
         print('404: ', v)
```

```python
cex_info.to_csv('cex_info.csv') 
```
