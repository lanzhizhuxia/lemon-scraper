```python
def get_market_ticker(code,page):
    url = '''https://dncapi.bqrank.net/api/v2/Coin/market_ticker?pagesize=100&code={CODE}&token=&tickertype=0&page={PAGE}&pair2=&webp=1'''.format(CODE=code,PAGE=page)
    strhtml = requests.get(url) #Get方式获取网页数据
    content = strhtml.text
    result = json.loads(content)
    data = result['data']['markets']
    code = result['code']
    cex = pd.DataFrame(data)
    return data
```
